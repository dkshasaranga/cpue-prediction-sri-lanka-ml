```python
# =============================================================================
# EVALUATING MACHINE LEARNING MODELS FOR PREDICTING CATCH PER UNIT EFFORT
# IN SRI LANKAN FISHERIES  ── OPTIMISED v4
# =============================================================================
# Changes from v3:
#   1. Removed Linear & Ridge  (explode with log-transform + grouped lags)
#   2. Added HuberRegressor     (robust linear baseline, safe with log target)
#   3. LOG_CAP applied before every expm1 back-transform (prevents explosions)
#   4. Non-zero CPUE performance table added (Phase 13-B)
#   5. All prior fixes retained: grouped lags, log target, zero-inflation
#      guard, tuned tree HPs, hard-decision Delta, correct Friedman test
# =============================================================================

# ─────────────────────────────────────────────────────────────────────────────
# PHASE 1 ── IMPORTS
# ─────────────────────────────────────────────────────────────────────────────
import warnings
warnings.filterwarnings("ignore")

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
from scipy.stats import friedmanchisquare

from sklearn.model_selection import TimeSeriesSplit
from sklearn.compose         import ColumnTransformer
from sklearn.pipeline        import Pipeline
from sklearn.impute          import SimpleImputer
from sklearn.preprocessing   import StandardScaler, OneHotEncoder
from sklearn.metrics         import mean_squared_error, mean_absolute_error, r2_score
from sklearn.linear_model    import HuberRegressor, LogisticRegression
from sklearn.ensemble        import (RandomForestRegressor, RandomForestClassifier,
                                      ExtraTreesRegressor)
from sklearn.inspection      import permutation_importance

import shap
from xgboost  import XGBRegressor,  XGBClassifier
from lightgbm import LGBMRegressor, LGBMClassifier
from catboost import CatBoostRegressor, CatBoostClassifier

os.makedirs("outputs", exist_ok=True)
SEED = 42
np.random.seed(SEED)

# ─────────────────────────────────────────────────────────────────────────────
# PHASE 2 ── LOAD DATASET
# ─────────────────────────────────────────────────────────────────────────────
print("=" * 70)
print("  CPUE PREDICTION – SRI LANKAN FISHERIES  (Optimised v4)")
print("=" * 70)

df = pd.read_csv("final_data_set.csv")
print(f"\n[INFO] Raw dataset : {df.shape}")

# ─────────────────────────────────────────────────────────────────────────────
# PHASE 3 ── CPUE TARGET & ZERO-INFLATION CHECK
# ─────────────────────────────────────────────────────────────────────────────
df["CPUE"] = (df["Catch"] / (df["True_Fishing_Time"] + 0.0001)).clip(lower=0)

zero_pct  = (df["CPUE"] == 0).mean() * 100
nonzero_n = (df["CPUE"] > 0).sum()
print(f"[INFO] Zero CPUE     : {zero_pct:.1f}%  ({(df['CPUE']==0).sum():,} rows)")
print(f"[INFO] Non-zero CPUE : {100-zero_pct:.1f}%  ({nonzero_n:,} rows)")

USE_DELTA = zero_pct >= 20.0
print(f"[INFO] Delta models  : {'YES  (zeros >= 20%)' if USE_DELTA else 'NO   (zeros < 20%)'}")

# ─────────────────────────────────────────────────────────────────────────────
# PHASE 4 ── DATE & TEMPORAL FEATURES
# ─────────────────────────────────────────────────────────────────────────────
df["landing_date"] = pd.to_datetime(df["landing_date"])
df["Month_Num"]    = df["landing_date"].dt.month
df["Year"]         = df["landing_date"].dt.year

def monsoon(m):
    if   m in [5,6,7,8,9]: return "SouthWest"
    elif m in [12,1,2]:     return "NorthEast"
    else:                   return "Inter"

df["Monsoon"] = df["Month_Num"].apply(monsoon)

# ─────────────────────────────────────────────────────────────────────────────
# PHASE 5 ── EFFORT & DERIVED FEATURES
# ─────────────────────────────────────────────────────────────────────────────
df["Effort_Efficiency"] = df["True_Fishing_Time"] / (df["Total_Fishing_Time"] + 0.0001)
df["Net_Density"]       = df["No_of_net_pieces"]  / (df["True_Fishing_Time"]  + 0.0001)
df["air_temp_C"]        = df["air_temp_mean_K"] - 273.15

# ─────────────────────────────────────────────────────────────────────────────
# PHASE 6 ── GROUPED LAG FEATURES  (per District x Gear_Boat_Code)
#   shift(1) BEFORE rolling  →  no same-row leakage
# ─────────────────────────────────────────────────────────────────────────────
df = df.sort_values(["District", "Gear_Boat_Code", "landing_date"]).reset_index(drop=True)

grp = df.groupby(["District", "Gear_Boat_Code"])

df["CPUE_Lag1"]  = grp["CPUE"].shift(1)
df["CPUE_Lag3"]  = grp["CPUE"].shift(1).transform(
    lambda x: x.rolling(3,  min_periods=1).mean())
df["CPUE_Lag12"] = grp["CPUE"].shift(1).transform(
    lambda x: x.rolling(12, min_periods=3).mean())
df["SST_Lag1"]   = grp["sst"].shift(1)
df["SST_Lag3"]   = grp["sst"].shift(1).transform(
    lambda x: x.rolling(3,  min_periods=1).mean())

df = df.dropna(subset=["CPUE_Lag1","CPUE_Lag3","CPUE_Lag12"]).reset_index(drop=True)
print(f"[INFO] After grouped lags : {df.shape}")

# ─────────────────────────────────────────────────────────────────────────────
# LOG CAP  – computed ONCE from full training data
#   Any log-space prediction above this cap is clipped before expm1
#   to prevent catastrophic back-transform explosion in linear models
# ─────────────────────────────────────────────────────────────────────────────
LOG_CAP = np.log1p(df["CPUE"].quantile(0.999))
print(f"[INFO] Log prediction cap : {LOG_CAP:.4f}  "
      f"(CPUE 99.9th pct = {df['CPUE'].quantile(0.999):.2f})")

# ─────────────────────────────────────────────────────────────────────────────
# PHASE 7 ── EDA
# ─────────────────────────────────────────────────────────────────────────────
print("\n[INFO] EDA visualisations ...")

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle("EDA - Sri Lankan Fisheries CPUE", fontsize=15, fontweight="bold")

# Zero inflation
zc = pd.Series({"Zero (=0)": (df["CPUE"]==0).sum(),
                 "Non-zero (>0)": (df["CPUE"]>0).sum()})
zc.plot(kind="bar", ax=axes[0,0], color=["#E76F51","#2A9D8F"], edgecolor="black")
axes[0,0].set_title(f"Zero Inflation  ({zero_pct:.1f}% zeros)")
axes[0,0].tick_params(axis="x", rotation=0)
for b in axes[0,0].patches:
    axes[0,0].text(b.get_x()+b.get_width()/2, b.get_height()+0.5,
                   f"{int(b.get_height()):,}", ha="center", va="bottom", fontsize=9)

# Raw CPUE
axes[0,1].hist(df["CPUE"], bins=80, color="#264653", edgecolor="white", linewidth=0.3)
axes[0,1].set_title("CPUE Distribution (Raw)"); axes[0,1].set_xlabel("CPUE")

# Log CPUE
axes[0,2].hist(np.log1p(df["CPUE"]), bins=80, color="#E9C46A",
               edgecolor="white", linewidth=0.3)
axes[0,2].set_title("log(CPUE+1) Distribution"); axes[0,2].set_xlabel("log(CPUE+1)")

# Annual trend
annual = df.groupby("Year")["CPUE"].mean()
axes[1,0].plot(annual.index, annual.values, marker="o", color="#E76F51", linewidth=2)
axes[1,0].fill_between(annual.index, annual.values, alpha=0.15, color="#E76F51")
axes[1,0].set_title("Annual Mean CPUE"); axes[1,0].set_xlabel("Year")

# Monthly seasonality
monthly = df.groupby("Month_Num")["CPUE"].mean()
axes[1,1].bar(monthly.index, monthly.values, color="#2A9D8F", edgecolor="black")
axes[1,1].set_title("Monthly Mean CPUE")
axes[1,1].set_xticks(range(1,13))
axes[1,1].set_xticklabels(["J","F","M","A","M","J","J","A","S","O","N","D"])

# Monsoon boxplot
df[df["Monsoon"].isin(["SouthWest","Inter","NorthEast"])].boxplot(
    column="CPUE", by="Monsoon", ax=axes[1,2], grid=False)
axes[1,2].set_title("CPUE by Monsoon")
plt.sca(axes[1,2]); plt.title("CPUE by Monsoon")

plt.tight_layout()
plt.savefig("outputs/01_eda.png", dpi=150, bbox_inches="tight")
plt.show()
print("[SAVED] outputs/01_eda.png")

# Correlation heatmap
num_eda = df.select_dtypes(include=np.number).drop(
    columns=["Catch","Year"], errors="ignore")
plt.figure(figsize=(16,12))
mask = np.triu(np.ones_like(num_eda.corr(), dtype=bool))
sns.heatmap(num_eda.corr(), mask=mask, cmap="coolwarm",
            annot=True, fmt=".2f", annot_kws={"size":6},
            linewidths=0.4, square=True)
plt.title("Correlation Matrix", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("outputs/02_correlation.png", dpi=150, bbox_inches="tight")
plt.show()
print("[SAVED] outputs/02_correlation.png")

# ─────────────────────────────────────────────────────────────────────────────
# PHASE 8 ── VALIDATED FEATURE SETS
#   Validated by 3-filter method:
#     Pearson |r| > 0.05  AND  MI > 0.01  AND  mean|SHAP| > 0.07
#   Removed: Catch (leakage), lon (zero variance), Year (r=0.008),
#            Quarter (SHAP=0), BoatType (MI=0, SHAP=0),
#            tp_weather_mean (fails 2/3), favorability_score (MI=0)
# ─────────────────────────────────────────────────────────────────────────────
NUM_FEATURES = [
    # CPUE autocorrelation — grouped per fishing unit
    "CPUE_Lag3", "CPUE_Lag1", "CPUE_Lag12",
    # Fishing effort
    "True_Fishing_Time", "Total_Fishing_Time",
    "No_of_net_pieces",  "Engine_Size",
    "Effort_Efficiency", "Net_Density",
    # Spatial
    "lat",
    # Ocean & weather
    "sst", "sst_weather_mean",
    "SST_Lag1", "SST_Lag3",
    "air_temp_C", "pressure_mean_hPa",
    "tp_weather_max",
    # Temporal
    "Month_Num",
]

CAT_FEATURES = [
    "District",       # MI=0.167
    "Gear_Boat_Code", # MI=0.075
    "Monsoon",        # SHAP=0.104
]

NUM_FEATURES = [c for c in NUM_FEATURES if c in df.columns]
CAT_FEATURES = [c for c in CAT_FEATURES if c in df.columns]
ALL_FEATURES = NUM_FEATURES + CAT_FEATURES

TARGET = "CPUE"
y_log  = np.log1p(df[TARGET])
X      = df[ALL_FEATURES]

print(f"\n[INFO] Features : {len(ALL_FEATURES)}  "
      f"(num={len(NUM_FEATURES)}, cat={len(CAT_FEATURES)})")
print(f"[INFO] Samples  : {len(X)}")
print(f"[INFO] Target   : log(CPUE+1) -> back-transformed to CPUE for metrics")

# ─────────────────────────────────────────────────────────────────────────────
# PHASE 9 ── PREPROCESSOR & CV SPLITTER
# ─────────────────────────────────────────────────────────────────────────────
num_pipe = Pipeline([
    ("imp",    SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
])
cat_pipe = Pipeline([
    ("imp", SimpleImputer(strategy="most_frequent")),
    ("ohe", OneHotEncoder(handle_unknown="ignore", sparse_output=False)),
])
preprocessor = ColumnTransformer([
    ("num", num_pipe, NUM_FEATURES),
    ("cat", cat_pipe, CAT_FEATURES),
], remainder="drop")

tscv = TimeSeriesSplit(n_splits=5)

# ── Visualise the TimeSeriesSplit folds ─────────────────────────────────────
fig, ax = plt.subplots(figsize=(14, 5))
n_samples = len(X)

for i, (tr_idx, te_idx) in enumerate(tscv.split(X)):
    ax.scatter(tr_idx, [i]*len(tr_idx), c="#2A9D8F", s=4, marker="s", label="Train" if i==0 else None)
    ax.scatter(te_idx, [i]*len(te_idx), c="#E76F51", s=4, marker="s", label="Test"  if i==0 else None)

ax.set_yticks(range(tscv.n_splits))
ax.set_yticklabels([f"Fold {i+1}" for i in range(tscv.n_splits)])
ax.set_xlabel("Sample index (chronological order)")
ax.set_title(f"TimeSeriesSplit Cross-Validation Scheme  ({tscv.n_splits} folds, n={n_samples:,})",
             fontsize=13, fontweight="bold")
ax.legend(loc="upper left", fontsize=10, markerscale=3)
ax.set_xlim(0, n_samples)

plt.tight_layout()
plt.savefig("outputs/00_timeseries_split.png", dpi=150, bbox_inches="tight")
plt.show()
print("[SAVED] outputs/00_timeseries_split.png")

# ─────────────────────────────────────────────────────────────────────────────
# PHASE 10 ── TWO-STAGE DELTA FUNCTION
#   Hard binary decision (prob >= threshold) instead of soft multiplication
#   LOG_CAP applied before every expm1 to prevent explosion
# ─────────────────────────────────────────────────────────────────────────────
def delta_predict(classifier, regressor, X_tr_p, y_tr_log, X_te_p,
                  threshold=0.40):
    """
    Stage-1 : P(CPUE > 0) classifier on original-scale binary labels
    Stage-2 : log(CPUE+1) regressor on positive-only rows
    Final   : hard binary mask x back-transformed CPUE
    """
    y_orig  = np.expm1(y_tr_log)
    y_class = (y_orig > 0).astype(int)

    # Skip classifier if almost no zeros
    if y_class.mean() > 0.98:
        pos_mask = y_orig > 0
        regressor.fit(X_tr_p[pos_mask], y_tr_log[pos_mask])
        log_hat = np.clip(regressor.predict(X_te_p), 0, LOG_CAP)
        return np.expm1(log_hat)

    classifier.fit(X_tr_p, y_class)
    prob        = classifier.predict_proba(X_te_p)[:, 1]
    binary_mask = (prob >= threshold).astype(float)

    pos_mask = y_orig > 0
    regressor.fit(X_tr_p[pos_mask], y_tr_log[pos_mask])
    log_hat = np.clip(regressor.predict(X_te_p), 0, LOG_CAP)

    return binary_mask * np.expm1(log_hat)

# ─────────────────────────────────────────────────────────────────────────────
# PHASE 11 ── MODEL DEFINITIONS
#   Linear & Ridge REMOVED — explode on log+grouped lag combination
#   HuberRegressor added as robust linear baseline
#   Tree hyperparameters tuned for time-series (depth/leaf regularisation)
# ─────────────────────────────────────────────────────────────────────────────
direct_models = {

    # Robust linear baseline (safe with log transform)
    "Huber"   : HuberRegressor(epsilon=1.35, max_iter=300),

    "RF"      : RandomForestRegressor(
                    n_estimators=500, max_depth=12,
                    min_samples_leaf=10, max_features=0.6,
                    random_state=SEED, n_jobs=-1),

    "ET"      : ExtraTreesRegressor(
                    n_estimators=500, max_depth=12,
                    min_samples_leaf=10, max_features=0.6,
                    random_state=SEED, n_jobs=-1),

    "XGB"     : XGBRegressor(
                    n_estimators=500, learning_rate=0.03,
                    max_depth=5, subsample=0.8,
                    colsample_bytree=0.7, min_child_weight=10,
                    reg_alpha=0.1, reg_lambda=1.0,
                    random_state=SEED, verbosity=0, eval_metric="rmse"),

    "LGBM"    : LGBMRegressor(
                    n_estimators=500, learning_rate=0.03,
                    num_leaves=31, min_child_samples=20,
                    subsample=0.8, colsample_bytree=0.7,
                    reg_alpha=0.1, reg_lambda=1.0,
                    random_state=SEED, verbose=-1),

    "CatBoost": CatBoostRegressor(
                    iterations=500, learning_rate=0.03,
                    depth=6, l2_leaf_reg=3.0,
                    random_seed=SEED, verbose=0),
}

delta_models = {}
if USE_DELTA:
    delta_models = {
        "Logistic_RF"  : (LogisticRegression(max_iter=500, random_state=SEED),
                          RandomForestRegressor(n_estimators=300, max_depth=10,
                                                min_samples_leaf=10,
                                                random_state=SEED, n_jobs=-1)),
        "Logistic_XGB" : (LogisticRegression(max_iter=500, random_state=SEED),
                          XGBRegressor(n_estimators=300, learning_rate=0.03,
                                       max_depth=5, min_child_weight=10,
                                       random_state=SEED, verbosity=0)),
        "RF_RF"        : (RandomForestClassifier(n_estimators=200,
                                                  random_state=SEED, n_jobs=-1),
                          RandomForestRegressor(n_estimators=300, max_depth=10,
                                                min_samples_leaf=10,
                                                random_state=SEED, n_jobs=-1)),
        "RF_XGB"       : (RandomForestClassifier(n_estimators=200,
                                                  random_state=SEED, n_jobs=-1),
                          XGBRegressor(n_estimators=300, learning_rate=0.03,
                                       max_depth=5, min_child_weight=10,
                                       random_state=SEED, verbosity=0)),
        "XGB_XGB"      : (XGBClassifier(n_estimators=200, learning_rate=0.05,
                                         max_depth=4, random_state=SEED,
                                         verbosity=0, eval_metric="logloss"),
                          XGBRegressor(n_estimators=300, learning_rate=0.03,
                                       max_depth=5, min_child_weight=10,
                                       random_state=SEED, verbosity=0)),
        "LGBM_LGBM"    : (LGBMClassifier(n_estimators=200, learning_rate=0.05,
                                           num_leaves=31, random_state=SEED,
                                           verbose=-1),
                          LGBMRegressor(n_estimators=300, learning_rate=0.03,
                                        num_leaves=31, min_child_samples=20,
                                        random_state=SEED, verbose=-1)),
        "CAT_CAT"      : (CatBoostClassifier(iterations=200, learning_rate=0.05,
                                              depth=5, random_seed=SEED, verbose=0),
                          CatBoostRegressor(iterations=300, learning_rate=0.03,
                                             depth=6, random_seed=SEED, verbose=0)),
    }
else:
    print("[INFO] Delta models SKIPPED — zero inflation below 20% threshold\n")

# ─────────────────────────────────────────────────────────────────────────────
# PHASE 12 ── CROSS-VALIDATION LOOP
#   Train  : log(CPUE+1)
#   Predict: clip log prediction to LOG_CAP then expm1
#   Metrics: reported on original CPUE scale
# ─────────────────────────────────────────────────────────────────────────────
print("\n" + "=" * 70)
print("  CROSS-VALIDATION  (5-fold TimeSeriesSplit | log-scale training)")
print("=" * 70)

def evaluate_orig(y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae  = mean_absolute_error(y_true, y_pred)
    r2   = r2_score(y_true, y_pred)
    return rmse, mae, r2

# ── 12-A  Direct models ───────────────────────────────────────────────────
direct_results   = {}
direct_oof_store = {}

for name, model in direct_models.items():
    fold_rmse, fold_mae, fold_r2 = [], [], []
    oof_preds, oof_actuals = [], []

    pipe = Pipeline([("pre", preprocessor), ("model", model)])

    for tr_idx, te_idx in tscv.split(X):
        X_tr,   X_te   = X.iloc[tr_idx],     X.iloc[te_idx]
        y_tr_l, y_te_l = y_log.iloc[tr_idx], y_log.iloc[te_idx]
        y_te_orig      = np.expm1(y_te_l.values)

        pipe.fit(X_tr, y_tr_l)
        log_hat    = np.clip(pipe.predict(X_te), 0, LOG_CAP)   # ← cap applied here
        preds_orig = np.clip(np.expm1(log_hat), 0, None)

        rmse, mae, r2 = evaluate_orig(y_te_orig, preds_orig)
        fold_rmse.append(rmse); fold_mae.append(mae); fold_r2.append(r2)
        oof_preds.extend(preds_orig); oof_actuals.extend(y_te_orig)

    direct_results[name] = {
        "RMSE_mean": np.mean(fold_rmse), "RMSE_std": np.std(fold_rmse),
        "MAE_mean" : np.mean(fold_mae),  "MAE_std" : np.std(fold_mae),
        "R2_mean"  : np.mean(fold_r2),   "R2_std"  : np.std(fold_r2),
        "fold_rmse": fold_rmse,
        "oof_preds": np.array(oof_preds), "oof_actuals": np.array(oof_actuals),
    }
    direct_oof_store[name] = fold_rmse
    print(f"  [Direct] {name:12s}  "
          f"RMSE={np.mean(fold_rmse):.4f}+-{np.std(fold_rmse):.4f}  "
          f"MAE={np.mean(fold_mae):.4f}  R2={np.mean(fold_r2):.4f}")

# ── 12-B  Delta models ────────────────────────────────────────────────────
delta_results   = {}
delta_oof_store = {}

for name, (clf, reg) in delta_models.items():
    fold_rmse, fold_mae, fold_r2 = [], [], []
    oof_preds, oof_actuals = [], []

    for tr_idx, te_idx in tscv.split(X):
        X_tr_raw, X_te_raw = X.iloc[tr_idx], X.iloc[te_idx]
        y_tr_l, y_te_l     = y_log.iloc[tr_idx], y_log.iloc[te_idx]
        y_te_orig          = np.expm1(y_te_l.values)

        X_tr_p = preprocessor.fit_transform(X_tr_raw, y_tr_l)
        X_te_p = preprocessor.transform(X_te_raw)

        preds_orig = np.clip(
            delta_predict(clf, reg, X_tr_p, y_tr_l.values, X_te_p), 0, None)

        rmse, mae, r2 = evaluate_orig(y_te_orig, preds_orig)
        fold_rmse.append(rmse); fold_mae.append(mae); fold_r2.append(r2)
        oof_preds.extend(preds_orig); oof_actuals.extend(y_te_orig)

    delta_results[name] = {
        "RMSE_mean": np.mean(fold_rmse), "RMSE_std": np.std(fold_rmse),
        "MAE_mean" : np.mean(fold_mae),  "MAE_std" : np.std(fold_mae),
        "R2_mean"  : np.mean(fold_r2),   "R2_std"  : np.std(fold_r2),
        "fold_rmse": fold_rmse,
        "oof_preds": np.array(oof_preds), "oof_actuals": np.array(oof_actuals),
    }
    delta_oof_store[name] = fold_rmse
    print(f"  [Delta]  {name:14s}  "
          f"RMSE={np.mean(fold_rmse):.4f}+-{np.std(fold_rmse):.4f}  "
          f"MAE={np.mean(fold_mae):.4f}  R2={np.mean(fold_r2):.4f}")

# ─────────────────────────────────────────────────────────────────────────────
# PHASE 13-A ── RESULTS TABLE & COMPARISON PLOTS
# ─────────────────────────────────────────────────────────────────────────────
def build_df(d, label):
    return pd.DataFrame([{
        "Model"    : n,    "Type"     : label,
        "RMSE_mean": round(v["RMSE_mean"],4), "RMSE_std": round(v["RMSE_std"],4),
        "MAE_mean" : round(v["MAE_mean"], 4), "MAE_std" : round(v["MAE_std"], 4),
        "R2_mean"  : round(v["R2_mean"],  4), "R2_std"  : round(v["R2_std"],  4),
    } for n, v in d.items()])

direct_df = build_df(direct_results, "Direct")
delta_df  = build_df(delta_results,  "Delta")
all_df    = pd.concat([direct_df, delta_df]).sort_values(
    "RMSE_mean").reset_index(drop=True)

print("\n-- Direct Models --")
print(direct_df.to_string(index=False))
if not delta_df.empty:
    print("\n-- Delta Models --")
    print(delta_df.to_string(index=False))
print("\n-- All Models Ranked by RMSE --")
print(all_df.to_string(index=False))
all_df.to_csv("outputs/model_results.csv", index=False)
print("\n[SAVED] outputs/model_results.csv")

# Bar chart comparison
from matplotlib.patches import Patch
palette = {"Direct": "#2A9D8F", "Delta": "#E76F51"}
leg     = [Patch(facecolor="#2A9D8F", label="Direct"),
           Patch(facecolor="#E76F51", label="Delta")]

fig, axes = plt.subplots(1, 3, figsize=(20, 7))
fig.suptitle("Model Comparison - CPUE Prediction (v4: log-transform + grouped lags + LOG_CAP)",
             fontsize=12, fontweight="bold")

for ax, metric, label in zip(axes,
    ["RMSE_mean","MAE_mean","R2_mean"],
    ["RMSE (lower is better)","MAE (lower is better)","R2 (higher is better)"]):
    colors = [palette[t] for t in all_df["Type"]]
    bars   = ax.bar(all_df["Model"], all_df[metric], color=colors,
                    edgecolor="black", linewidth=0.6)
    best_i = all_df[metric].idxmin() if "R2" not in metric else all_df[metric].idxmax()
    bars[best_i].set_edgecolor("gold"); bars[best_i].set_linewidth(3)
    ax.set_title(label, fontsize=11)
    ax.set_xticklabels(all_df["Model"], rotation=45, ha="right", fontsize=8)
    ax.set_ylabel(metric.replace("_mean",""))
    ax.legend(handles=leg, fontsize=8)

plt.tight_layout()
plt.savefig("outputs/03_model_comparison.png", dpi=150, bbox_inches="tight")
plt.show()
print("[SAVED] outputs/03_model_comparison.png")

# CV RMSE boxplot
fold_rows = []
for n, v in {**direct_results, **delta_results}.items():
    for r in v["fold_rmse"]:
        fold_rows.append({"Model": n, "Fold_RMSE": r})
fold_df = pd.DataFrame(fold_rows)

plt.figure(figsize=(16, 6))
sns.boxplot(data=fold_df, x="Model", y="Fold_RMSE",
            order=all_df["Model"].tolist(), palette="Set2")
plt.xticks(rotation=45, ha="right")
plt.title("CV RMSE Distribution per Model  (5 folds)", fontsize=12, fontweight="bold")
plt.tight_layout()
plt.savefig("outputs/04_cv_boxplot.png", dpi=150, bbox_inches="tight")
plt.show()
print("[SAVED] outputs/04_cv_boxplot.png")

# ─────────────────────────────────────────────────────────────────────────────
# PHASE 13-B ── NON-ZERO CPUE PERFORMANCE TABLE
#   With 54% zeros, overall R2 is deflated by easy zero predictions.
#   This table shows TRUE predictive power on actual catch events.
# ─────────────────────────────────────────────────────────────────────────────
print("\n" + "=" * 70)
print("  PERFORMANCE ON NON-ZERO CPUE ROWS ONLY  (actual catch events)")
print("=" * 70)

nz_rows = []
for name, v in {**direct_results, **delta_results}.items():
    mask = v["oof_actuals"] > 0
    if mask.sum() < 10:
        continue
    act  = v["oof_actuals"][mask]
    pred = v["oof_preds"][mask]
    nz_rows.append({
        "Model"    : name,
        "Type"     : "Direct" if name in direct_results else "Delta",
        "RMSE"     : round(np.sqrt(mean_squared_error(act, pred)), 4),
        "MAE"      : round(mean_absolute_error(act, pred), 4),
        "R2"       : round(r2_score(act, pred), 4),
        "N_nonzero": int(mask.sum()),
    })

nz_df = pd.DataFrame(nz_rows).sort_values("RMSE").reset_index(drop=True)
print(nz_df.to_string(index=False))
nz_df.to_csv("outputs/nonzero_performance.csv", index=False)
print("[SAVED] outputs/nonzero_performance.csv")

# ─────────────────────────────────────────────────────────────────────────────
# PHASE 13-C ── STAGE-1 CLASSIFIER PERFORMANCE  (Delta models only)
#   With 54% zeros the classifier is doing critical work.
#   This section evaluates how well it separates zero vs non-zero catch.
# ─────────────────────────────────────────────────────────────────────────────
if USE_DELTA:
    from sklearn.metrics import classification_report, roc_auc_score

    print("\n" + "=" * 70)
    print("  STAGE-1 CLASSIFIER PERFORMANCE  (best Delta model: CAT_CAT)")
    print("=" * 70)

    splits_tmp          = list(tscv.split(X))
    tr_idx_c, te_idx_c  = splits_tmp[-1]

    X_tr_c = preprocessor.fit_transform(X.iloc[tr_idx_c], y_log.iloc[tr_idx_c])
    X_te_c = preprocessor.transform(X.iloc[te_idx_c])

    y_tr_orig_c = np.expm1(y_log.iloc[tr_idx_c].values)
    y_te_orig_c = np.expm1(y_log.iloc[te_idx_c].values)

    y_tr_class  = (y_tr_orig_c > 0).astype(int)
    y_te_class  = (y_te_orig_c > 0).astype(int)

    # Use the CatBoost classifier from CAT_CAT
    clf_eval, _ = delta_models["CAT_CAT"]
    clf_eval.fit(X_tr_c, y_tr_class)

    y_pred_class = clf_eval.predict(X_te_c)
    y_pred_prob  = clf_eval.predict_proba(X_te_c)[:, 1]

    print(classification_report(
        y_te_class, y_pred_class,
        target_names=["Zero catch (0)", "Positive catch (>0)"],
        digits=4
    ))
    auc = roc_auc_score(y_te_class, y_pred_prob)
    print(f"  AUC-ROC : {auc:.4f}")

    # ROC curve
    from sklearn.metrics import roc_curve
    fpr, tpr, _ = roc_curve(y_te_class, y_pred_prob)
    plt.figure(figsize=(6, 5))
    plt.plot(fpr, tpr, color="#E76F51", linewidth=2,
             label=f"CAT_CAT classifier  (AUC = {auc:.4f})")
    plt.plot([0,1],[0,1], "k--", linewidth=1, label="Random")
    plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate")
    plt.title("ROC Curve - Stage-1 Classifier (Zero vs Positive Catch)")
    plt.legend(fontsize=10); plt.tight_layout()
    plt.savefig("outputs/11_roc_curve.png", dpi=150, bbox_inches="tight")
    plt.show()
    print("[SAVED] outputs/11_roc_curve.png")

    from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

    cm  = confusion_matrix(y_te_class, y_pred_class)
    fig, ax = plt.subplots(figsize=(6, 5))
    disp = ConfusionMatrixDisplay(
        confusion_matrix=cm,
        display_labels=["Zero catch (0)", "Positive catch (>0)"]
    )
    disp.plot(ax=ax, colorbar=True, cmap="Blues")
    ax.set_title(
        "Confusion Matrix — CAT_CAT Stage-1 Classifier\n"
        "(Last fold of TimeSeriesSplit  |  threshold=0.40)",
        fontsize=11, fontweight="bold"
    )
    ax.set_xlabel("Predicted Label", fontsize=10)
    ax.set_ylabel("True Label", fontsize=10)
    plt.tight_layout()
    plt.savefig("outputs/14_confusion_matrix_classifier.png",
                dpi=150, bbox_inches="tight")
    plt.show()
    print("[SAVED] outputs/14_confusion_matrix_classifier.png")
        
# ─────────────────────────────────────────────────────────────────────────────
# PHASE 14 ── BEST MODEL — RETRAIN ON LAST FOLD FOR DIAGNOSTICS
# ─────────────────────────────────────────────────────────────────────────────
best_row  = all_df.iloc[0]
best_name = best_row["Model"]
best_type = best_row["Type"]
print(f"\n[INFO] Best model : {best_name}  ({best_type})  "
      f"RMSE={best_row['RMSE_mean']:.4f}  R2={best_row['R2_mean']:.4f}")

splits              = list(tscv.split(X))
tr_idx_f, te_idx_f = splits[-1]
X_tr_f, X_te_f     = X.iloc[tr_idx_f], X.iloc[te_idx_f]
y_tr_lf, y_te_lf   = y_log.iloc[tr_idx_f], y_log.iloc[te_idx_f]
y_test_orig         = np.expm1(y_te_lf.values)

if best_type == "Direct":
    best_pipe = Pipeline([("pre", preprocessor),
                           ("model", direct_models[best_name])])
    best_pipe.fit(X_tr_f, y_tr_lf)
    log_hat        = np.clip(best_pipe.predict(X_te_f), 0, LOG_CAP)
    final_preds    = np.clip(np.expm1(log_hat), 0, None)
    X_te_proc      = best_pipe["pre"].transform(X_te_f)
    best_raw_model = best_pipe["model"]
else:
    clf, reg    = delta_models[best_name]
    X_tr_p      = preprocessor.fit_transform(X_tr_f, y_tr_lf)
    X_te_proc   = preprocessor.transform(X_te_f)
    final_preds = np.clip(
        delta_predict(clf, reg, X_tr_p, y_tr_lf.values, X_te_proc), 0, None)
    best_raw_model = reg

try:
    ohe_names     = (preprocessor.named_transformers_["cat"]["ohe"]
                     .get_feature_names_out(CAT_FEATURES).tolist())
    feature_names = NUM_FEATURES + ohe_names
except Exception:
    feature_names = [f"f{i}" for i in range(X_te_proc.shape[1])]

# ─────────────────────────────────────────────────────────────────────────────
# PHASE 15 ── RESIDUAL DIAGNOSTICS
# ─────────────────────────────────────────────────────────────────────────────
residuals = y_test_orig - final_preds

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle(f"Residual Diagnostics - {best_name}", fontsize=13, fontweight="bold")

axes[0].hist(residuals, bins=50, color="#264653", edgecolor="white", linewidth=0.3)
axes[0].axvline(0, color="red", linestyle="--")
axes[0].set_title("Residual Distribution"); axes[0].set_xlabel("Residual")

(osm, osr), (slope, intercept, _) = stats.probplot(residuals, dist="norm")
axes[1].scatter(osm, osr, s=10, alpha=0.6, color="#2A9D8F")
axes[1].plot(osm, slope*np.array(osm)+intercept, color="red", linewidth=2)
axes[1].set_title("Q-Q Plot")
axes[1].set_xlabel("Theoretical Quantiles"); axes[1].set_ylabel("Sample Quantiles")

axes[2].scatter(final_preds, residuals, s=10, alpha=0.4, color="#E76F51")
axes[2].axhline(0, color="black", linestyle="--")
axes[2].set_title("Residuals vs Fitted")
axes[2].set_xlabel("Fitted CPUE"); axes[2].set_ylabel("Residual")

plt.tight_layout()
plt.savefig("outputs/05_residuals.png", dpi=150, bbox_inches="tight")
plt.show()
print("[SAVED] outputs/05_residuals.png")

# Actual vs Predicted
lims = [min(y_test_orig.min(), final_preds.min())*0.95,
        max(y_test_orig.max(), final_preds.max())*1.05]
plt.figure(figsize=(7,6))
plt.scatter(y_test_orig, final_preds, s=12, alpha=0.4, color="#264653")
plt.plot(lims, lims, "r--", linewidth=1.5, label="Perfect prediction")
plt.xlabel("Actual CPUE"); plt.ylabel("Predicted CPUE")
plt.title(f"Actual vs Predicted - {best_name}  "
          f"(R2={r2_score(y_test_orig, final_preds):.4f})")
plt.legend(); plt.tight_layout()
plt.savefig("outputs/06_actual_vs_predicted.png", dpi=150, bbox_inches="tight")
plt.show()
print("[SAVED] outputs/06_actual_vs_predicted.png")

# ─────────────────────────────────────────────────────────────────────────────
# PHASE 16 ── SHAP ANALYSIS
# ─────────────────────────────────────────────────────────────────────────────
print("\n[INFO] SHAP analysis ...")
try:
    if hasattr(best_raw_model, "feature_importances_"):
        explainer   = shap.TreeExplainer(best_raw_model)
        shap_values = explainer.shap_values(X_te_proc)
    else:
        bg          = shap.sample(X_te_proc, 100, random_state=SEED)
        explainer   = shap.KernelExplainer(best_raw_model.predict, bg)
        shap_values = explainer.shap_values(X_te_proc)

    if isinstance(shap_values, list):
        shap_values = shap_values[1]

    shap_df = pd.DataFrame(shap_values, columns=feature_names)

    plt.figure()
    shap.summary_plot(shap_values, X_te_proc,
                      feature_names=feature_names, show=False, max_display=20)
    plt.title(f"SHAP Summary - {best_name}")
    plt.tight_layout()
    plt.savefig("outputs/07_shap_summary.png", dpi=150, bbox_inches="tight")
    plt.show()
    print("[SAVED] outputs/07_shap_summary.png")

    mean_shap = np.abs(shap_df).mean().sort_values(ascending=False).head(20)
    plt.figure(figsize=(10,6))
    mean_shap[::-1].plot(kind="barh", color="#2A9D8F", edgecolor="black")
    plt.title(f"Top 20 Features (mean |SHAP|) - {best_name}")
    plt.xlabel("mean |SHAP|")
    plt.tight_layout()
    plt.savefig("outputs/08_shap_bar.png", dpi=150, bbox_inches="tight")
    plt.show()
    print("[SAVED] outputs/08_shap_bar.png")

except Exception as e:
    print(f"[WARN] SHAP skipped : {e}")

# ─────────────────────────────────────────────────────────────────────────────
# PHASE 17 ── PERMUTATION IMPORTANCE
# ─────────────────────────────────────────────────────────────────────────────
print("\n[INFO] Permutation importance ...")
try:
    perm    = permutation_importance(
                  best_raw_model, X_te_proc, y_test_orig,
                  n_repeats=20, random_state=SEED,
                  scoring="neg_root_mean_squared_error")
    perm_df = pd.DataFrame({
        "Feature"   : feature_names,
        "Importance": perm.importances_mean,
        "Std"       : perm.importances_std,
    }).sort_values("Importance", ascending=False).head(20)

    print("\nTop-20 Permutation Importance:")
    print(perm_df.to_string(index=False))

    plt.figure(figsize=(10,7))
    plt.barh(perm_df["Feature"][::-1], perm_df["Importance"][::-1],
             xerr=perm_df["Std"][::-1], color="#E9C46A",
             edgecolor="black", capsize=3)
    plt.xlabel("Permutation Importance (mean delta RMSE)")
    plt.title(f"Permutation Feature Importance - {best_name}")
    plt.tight_layout()
    plt.savefig("outputs/09_permutation_importance.png", dpi=150, bbox_inches="tight")
    plt.show()
    print("[SAVED] outputs/09_permutation_importance.png")
except Exception as e:
    print(f"[WARN] Permutation importance skipped : {e}")

# ─────────────────────────────────────────────────────────────────────────────
# PHASE 18 ── TEMPORAL DRIFT
# ─────────────────────────────────────────────────────────────────────────────
df_test              = df.iloc[te_idx_f].copy().reset_index(drop=True)
df_test["Predicted"] = final_preds

yearly_rmse  = (df_test.groupby("Year")
                .apply(lambda g: np.sqrt(mean_squared_error(g["CPUE"], g["Predicted"])))
                .reset_index(name="RMSE"))
monthly_rmse = (df_test.groupby("Month_Num")
                .apply(lambda g: np.sqrt(mean_squared_error(g["CPUE"], g["Predicted"])))
                .reset_index(name="RMSE"))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle(f"Temporal Drift - {best_name}", fontsize=13, fontweight="bold")
axes[0].plot(yearly_rmse["Year"], yearly_rmse["RMSE"],
             marker="o", color="#E76F51", linewidth=2)
axes[0].set_title("Annual RMSE Drift"); axes[0].set_xlabel("Year"); axes[0].set_ylabel("RMSE")
axes[1].bar(monthly_rmse["Month_Num"], monthly_rmse["RMSE"],
            color="#2A9D8F", edgecolor="black")
axes[1].set_title("Monthly RMSE Pattern")
axes[1].set_xlabel("Month"); axes[1].set_ylabel("RMSE"); axes[1].set_xticks(range(1,13))
plt.tight_layout()
plt.savefig("outputs/10_temporal_drift.png", dpi=150, bbox_inches="tight")
plt.show()
print("[SAVED] outputs/10_temporal_drift.png")

# ─────────────────────────────────────────────────────────────────────────────
# PHASE 19 ── FRIEDMAN STATISTICAL TEST
#   Correct usage: 5 per-fold RMSE values per model (not per-observation errors)
# ─────────────────────────────────────────────────────────────────────────────
print("\n" + "=" * 70)
print("  FRIEDMAN STATISTICAL TEST  (per-fold RMSE, k models x 5 folds)")
print("=" * 70)

all_fold_rmse = {**direct_oof_store, **delta_oof_store}
error_matrix  = [np.array(v) for v in all_fold_rmse.values()]

if len(error_matrix) >= 3:
    stat, p_val = friedmanchisquare(*error_matrix)
    print(f"\n  Models compared       : {len(error_matrix)}")
    print(f"  Friedman chi2         : {stat:.4f}")
    print(f"  p-value               : {p_val:.6f}")
    if p_val < 0.05:
        print("  SIGNIFICANT — models differ statistically (p < 0.05)")
    else:
        print("  NOT SIGNIFICANT — no statistical difference detected (p >= 0.05)")
else:
    print("[WARN] Need >= 3 models.")

# ─────────────────────────────────────────────────────────────────────────────
# PHASE 20 ── FINAL SUMMARY
# ─────────────────────────────────────────────────────────────────────────────
print("\n" + "=" * 70)
print("  FINAL RANKED RESULTS  (all models, original CPUE scale)")
print("=" * 70)
print(all_df.to_string(index=True))

print("\n" + "=" * 70)
print("  NON-ZERO CPUE PERFORMANCE  (actual catch events only)")
print("=" * 70)
print(nz_df.to_string(index=True))

best_r  = all_df.iloc[0]
best_nz = nz_df.iloc[0] if not nz_df.empty else None

print(f"""
+----------------------------------------------------------------------+
|  BEST MODEL (overall)   : {best_r['Model']:<15} Type: {best_r['Type']:<8}         |
|  RMSE                   : {best_r['RMSE_mean']:<10.4f}                              |
|  MAE                    : {best_r['MAE_mean']:<10.4f}                              |
|  R2 (all rows)          : {best_r['R2_mean']:<10.4f}                              |""")
if best_nz is not None:
    print(f"|  R2 (non-zero rows)     : {best_nz['R2']:<10.4f}  <- true predictive power    |")
print(f"""|                                                                      |
|  Target scale  : Original CPUE  (trained on log(CPUE+1))            |
|  Lags          : Grouped by District x Gear_Boat_Code               |
|  Zero CPUE     : {zero_pct:.1f}%  -> Delta models {'included' if USE_DELTA else 'excluded'}                         |
|  LOG_CAP       : {LOG_CAP:.4f}  (prevents expm1 explosion)              |
+----------------------------------------------------------------------+
""")
print("[INFO] All figures saved in ./outputs/")
print("[DONE]")
```
